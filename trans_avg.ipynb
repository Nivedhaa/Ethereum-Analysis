```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import pandas as pd

# Open the file and read its contents into a list
with open('transactions_average.txt', 'r') as file:
    file_contents = file.read().splitlines()

# Extract the data from the list and split it into a list of values
data_string = file_contents[0].replace('[','').replace(']','').replace('"','')
data_list = data_string.split(',')

# Create a dictionary from the list of values
data_dict = {'Month':[], 'Average_transaction':[]}
for i in range(0, len(data_list), 2):
    data_dict['Month'].append(data_list[i])
    data_dict['Average_transaction'].append(float(data_list[i+1]))

# Create a DataFrame from the dictionary
df = pd.DataFrame(data_dict)

# Sort the DataFrame by month
df_sorted = df.sort_values(by='Month')

# Print the sorted DataFrame
print(df_sorted)
```

```python
# Bar Plot
plt.figure(figsize = (10, 10))
plt.bar(df_sorted['Month'], df_sorted['Average_transaction'])
plt.xticks(rotation = 90)
plt.xlabel('Month')
plt.ylabel('Average Transaction')
plt.title('Average Transaction per Month')
plt.show()
```

```python
# Line Plot
plt.figure(figsize = (10, 10))
plt.plot(df_sorted['Month'], df_sorted['Average_transaction'])
plt.xticks(rotation = 90)
plt.xlabel('Month')
plt.ylabel('Average Transaction')
plt.title('Average Transaction per Month')
plt.show()
```
