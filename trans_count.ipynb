```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('transactions_total.txt', 'r') as f:
    file_contents = f.read()
```

```python
data = eval(file_contents)
df = pd.DataFrame(data, columns=['Date', 'Value'])
df
```

```python
df['Date'] = df['Date'].str.replace("/",",")
df
```

```python
df = df.sort_values(by=['Date'])
```

```python
plt.figure(figsize=(10,10))
plt.bar(df["Date"],df['Value'])
plt.xlabel("Date")
plt.ylabel("Values")
plt.title("Transaction per month")
plt.xticks(rotation = 90)
plt.savefig("Transaction_per_month.png")
plt.show()
```
